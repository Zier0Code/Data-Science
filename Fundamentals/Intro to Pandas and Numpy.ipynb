## Pandas and Numpy

## 1. Creating a data frame from an array

#### 1.1 Option 1

```python
import pandas as pd
import numpy as np
```

```python
#creating an array
data = np.array([[1,3], [4,5], [5,7]])
```

```python
#creating DataFrame
pd.DataFrame(data, index=['row1','row2','row3'],columns=['col1','col2'])
```

#### 1.2 Option 2

```python
#creating an array with list shape
data = [[1,2],[3,4],[5,6]]
```

```python
#creating dataframe
df = pd.DataFrame(data, index=['row1','row2','row3'],columns=['col1','col2'])
```

```python
#showing dataframe
df
```

## Creating Dataframe from dictionary



```python
# lists used for this example
states=['California','Florida','New York','Texas']
population=[5646456,456465,456465,78998]
```

## 10 Sorting DataFrame

```python
# sort by lowest to highest value 
# ONE COLUMN!
# for descending value

# df_anime.sort_values(by='Best Anime', ascending=False)
df_anime.sort_values(by='Best Anime')
```

```python
# MULTIPLE COLUMN SORT
# Also to Update the Dataframe Easily
df_anime.sort_values(by=['Best Anime','Best Anime1'], ascending=False,inplace=True)
```

```python
df_anime
```

```python
# sort values with key function

df_anime.sort_values('Name',ascending=True, 
                   key=lambda col:col.str.lower()).head()
```
